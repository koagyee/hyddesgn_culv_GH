```python
# This python code is for Hydraulic Design of Culverts

# 1.0 Bed and Land Slope Calculation
X1A = float(input('Enter Highest Point in Catchment in feet'))
X2A = float(input('Enter Point within River Bed in feet'))
X3A = float(input('Enter Height at Crossing in feet'))
X4 = float(input('Enter Stream length at Crossing in meters'))
X5 = float(input('Enter Length of Stream in meters'))
# 1.1 The equations below convert X1F, X2F and X3F from feet to meters
X1 = X1A/3.28
X2 = X2A/3.28
X3 = X3A/3.28       
# 1.2 The equations S and R below calculate the Bed Slope and Slope of Land respectively
S = round((((X2-X3)/X4)*100),2)
R = round((((X1-X3)/X5)*100),2)
print('Bed Slope, S = ', S, '%')
print('Slope of Land, R = ', R, '%')
```

```python
# 2.0 Average Daily Rainfall Calculation
RFI = float(input('Enter Average Annual Rainfall in inches'))
# 2.1 The equation below converts M1I from inches to millimeters
RF = round(((RFI)*25.4),2)
print('Average Annual Rainfall, RF = ', RF, 'mm')
# 2.2 The equation below calculates the Average Daily Rainfall
RF10 = round(((0.033*RF)+82.7),2)
print('Average Daily Rainfall, RF10 = ', RF10, 'mm')
```

```python
#3.0 Design Parameters and Areal Reduction Factor Calculation
#3.1 Time of Base for Volta Region is 5hrs
#3.2 Time of Base for Central Region is 7hrs
#3.3 Slope Category for V/R and C/R is R4
#3.4 Infiltration Category for V/R and C/R is P3
#3.5 Runoff Coefficient for V/R and C/R is 0.40
#3.6 Mannings Coefficient of Roughness for V/R and C/R is 0.013 
Tb = float(input('Enter Time of Base in hours'))
```

```python
#4.0 Estimated Areal Reduction Factor and Design Discharge Calculation
CA = float(input('Enter Catchment Area in square kilometers'))
if CA > 1 and CA < 3:
   C10 = 1
   print('Areal Reduction Factor, C10 = ', C10)
elif CA > 15:
   import math
   C10 = round((1-(9*math.log10(10)-(0.042*RF)+152)*0.001*math.log10(CA)),2)
   print('Areal Reduction Factor, C10 = ', C10)
else:
   C10 = round((1.0245-0.0084*CA),2)
   print('Areal Reduction Factor, C10 = ', C10)
if CA < 120:
   Kp = 2.5
   print('Peak to Average Discharge Ratio, Kp = ', Kp)
elif CA >= 120 and CA <= 200:
   Kp = 3.12
   print('Peak to Average Discharge Ratio, Kp = ', Kp)
Kr = float(input('Enter Runoff Coefficient'))       
Q9 = (Kp * Kr * C10 * RF10 * CA)/(3.6 * Tb)
Q10 = round((Q9),2)
Q25 = round((1.15 * Q10),2)
print('Q10 = ', Q10, 'cubic meters per second')
print('Estimated Design Discharge, Q25 = ', Q25, 'cubic meters per second')
```

```python
#5.0 Determination of Theoretical Depth of Culvert
n = float(input('Enter Mannings Coefficient of Roughness'))
K1 = 5.6
K2 = 1.25
#5.1 K = K1/K2
K = 4.48
d = round(((Q25/(K*((S/100)**0.5)/n))**(1/2.22)),2)
print('Theoretical depth, d = ', d, 'meters')

#6.0 Determination of Theoretical Area of Culvert
A = round(((n*Q25)/((d**(2/3)/K2)*(S/100)**0.5)),2)
print('Theoretical Area, A = ', A, 'square meters')
```
