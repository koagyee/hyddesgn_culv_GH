```python
# This python code is for Hydraulic Design of Culverts in DFR, Ghana


def bed_slope(X,Y,Z):
     print('Bed Slope, S =', round((((X/3.28-Y/3.28)/Z)*100),2), '%')
    

bed_slope(350,340,1100)
    
def slope_land(U,V,Y):
     print('Slope of Land, R =', round((((U/3.28-Y/3.28)/V)*100),2), '%')
    

slope_land(550,3600,340)

 # X is point within River Bed in feet
 # Y is Height at Crossing in feet
 # Z is Stream Length at Crossing in meters
 # U is Highest Point within Catchment in feet
 # V is Length of Stream in meters
```

```python
def average_daily_rainfall(W):
    print('Average Daily Rainfall, RF10 =', round(((0.033*W*25.4)+82.7),2), 'mm')

average_daily_rainfall(45)
 # W is Average Annual Rainfall in inches

#3.0 Design Parameters and Areal Reduction Factor Calculation
#3.1 Time of Base for Volta Region is 5hrs
#3.2 Time of Base for Central Region is 7hrs
#3.3 Slope Category for V/R and C/R is R4
#3.4 Infiltration Category for V/R and C/R is P3
#3.5 Runoff Coefficient for V/R and C/R is 0.40
#3.6 Mannings Coefficient of Roughness for V/R and C/R is 0.013
#3.7 C10 is Areal Reduction Factor
#3.8 Kp is Peak to Average Discharge Ratio
```

```python
def Areal_Rdn_Factor(B):
 if B > 1 and B < 3:
     print('Areal Reduction Factor, C10 =', 1)
 elif B > 15:
   import math
   print('Areal Reduction Factor, C10 =', round((1-(9*math.log10(10)-(0.042*W*25.4)+152)*0.001*math.log10(B)),2))
 else:
   print('Areal Reduction Factor, C10 =', round((1.0245-0.0084*B),2))
 if B < 120:
   print('Peak to Average Discharge Ratio, Kp =', 2.5) 
 elif B >= 120 and B <= 200:
   print('Peak to Average Discharge Ratio, Kp =', 3.12)

Areal_Rdn_Factor(1.7)
```

```python
def Est_Dsgn_Discharge(Kp,C,C10,W,B,A):

     print('Estimated Design Discharge, Q25 =',round((1.15 * (Kp * C * C10 * ((0.033*W*25.4)+82.7) * B)/(3.6 * A)),2) , 'cubic meters per second')    

Est_Dsgn_Discharge(2.5,0.4,1,45,1.7,7)

# A, Time of Base in hours is 7 for Central Region
# B is Catchment Area in square kilometers
# C, Runoff Coefficient is 0.40 for V/R and C/R
# W is Average Annual Rainfall in inches
```

```python
def Theo_depth(n,Q25,S):

    print('Theoretical depth, d =', round(((Q25/(4.48*((S/100)**0.5)/n))**(1/2.22)),2), 'meters')

Theo_depth(0.013,9.34,0.28)

# n, Mannings Coefficient of Roughness is 0.013

def Theo_Area(n,Q25,d,S):

    print('Theoretical Area, A =', round(((n*Q25)/((d**(2/3)/1.25)*(S/100)**0.5)),2), 'square meters')

Theo_Area(0.013,9.34,0.74,0.28)
```
